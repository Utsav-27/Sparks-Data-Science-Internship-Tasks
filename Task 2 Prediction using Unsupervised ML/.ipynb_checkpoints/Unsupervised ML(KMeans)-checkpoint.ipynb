```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
iris = sns.load_dataset('Iris').drop("species", axis=1)
iris.head()
```

```python
x = iris.iloc[:, [0,1,2,3]].values
```

```python
sse = []
for k in range(1,11):
    km = KMeans(n_clusters=k)
    km.fit(x)
    sse.append(km.inertia_)
sse
```

```python
plt.xlabel("Total Clusters Formed")
plt.ylabel("Sum of Squared Error")
plt.plot(range(1,11), sse)
```

```python
### Taking n = 3
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(x)
y_predicted
```

```python
iris['cluster'] = y_predicted
iris
```

## Observing Clusters on Sepal Length and Sepal Width

```python
df1 = iris[iris['cluster'] == 0]
df2 = iris[iris['cluster'] == 1]
df3 = iris[iris['cluster'] == 2]
plt.scatter(df1['sepal_length'], df1['sepal_width'], color="red", label="Iris-setosa")
plt.scatter(df2['sepal_length'], df2['sepal_width'], color="blue", label="Iris-versicolor")
plt.scatter(df3['sepal_length'], df3['sepal_width'], color="green", label="Iris-virginica")
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:,1], color="yellow", marker="*", label="centroid")
plt.legend(bbox_to_anchor = (1,1),loc="upper left")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

## Observing Clusters on Petal Length and Petal Width

```python
plt.scatter(df1['petal_length'], df1['petal_width'], color="red", label="Iris-setosa")
plt.scatter(df2['petal_length'], df2['petal_width'], color="blue", label="Iris-versicolor")
plt.scatter(df3['petal_length'], df3['petal_width'], color="green", label="Iris-virginica")
plt.scatter(km.cluster_centers_[:, 2], km.cluster_centers_[:,3], color="yellow", marker="*", label="centroid")
plt.legend(bbox_to_anchor = (1,1),loc="upper left")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.show()
```
