### Visualization

#### Which Sub-Category did the best sales?

```python
plt.figure(figsize=(20,9))
plt.bar(df["Sub-Category"], df["Sales"])
plt.show()
```

#### Which Category did the best Sales?

```python
plt.bar(df["Category"], df["Sales"])
plt.show()
```

#### Which category gave the best profit?

```python
plt.bar(df["Category"], df["Profit"])
plt.show()
```

#### Which Sub-category gave the best profit?

```python
plt.figure(figsize=(20,9))
plt.bar(df["Sub-Category"], df["Profit"])
plt.show()
```

#### Check for the sub-category which have high-discounts

```python
plt.figure(figsize=(20,12))
plt.bar(df["Sub-Category"], df["Discount"])
plt.show()
```

#### Which state have highest supplies?

```python
plt.figure(figsize=(20,9))
sns.countplot(x=df['State'])
plt.xticks(rotation=90)
plt.show()
```

#### Which sub-category have highest supply?

```python
plt.figure(figsize=(20,9))
sns.countplot(x=df['Sub-Category'])
plt.xticks(rotation=90)
plt.show()
```

#### Which region have high supply?

```python
sns.countplot(x=df['Region'])
```

#### Which Segment have highest supply?

```python
sns.countplot(x=df['Segment'])
```

#### Here, we can see that, higher the discounts lower the Profit
